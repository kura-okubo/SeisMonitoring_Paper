```python
import pandas as pd
```

```python
df = pd.read_csv('./shelly2017catalog.csv', comment='%', header=0)
df
```

```python
IDs = df["ID"].unique()
len(IDs)
```

```python
# df_family["latitude"].iloc[0]
```

```python
df_sseloc=pd.DataFrame()
for ID in IDs:
    df_family=df[df.ID == ID]
    # check if all latlon is same within the family
    if (len(df_family["latitude"].unique()) == 1) and (len(df_family["longitude"].unique()) == 1):
#         print([df_family["latitude"].iloc[0], df_family["longitude"].iloc[0]])
        df_sseloc = df_sseloc.append({'latitude': df_family["latitude"].iloc[0], 
                                      'longitude': df_family["longitude"].iloc[0],
                                      'depth': df_family["depth"].iloc[0]}, ignore_index=True)
        continue
```

```python
df_sseloc.latitude=df_sseloc.latitude.astype(float)
df_sseloc.longitude=df_sseloc.longitude.astype(float)
df_sseloc.depth=df_sseloc.depth.astype(float)
```

```python
df_sseloc.to_csv('gmt_shelly2017ssefamily.txt', columns=['longitude', 'latitude'], index=None, header=None, sep=",")
```

```python
df_sseloc.to_csv('crosssection_shelly2017ssefamily.txt', columns=['longitude', 'latitude', 'depth'], index=None, header=None, sep=",")
```
